```python
def midi_to_2d(midi):
    d = []
    i = 0
    while i < len(midi) - 1:
        p = midi[i]
        l = 1
        while midi[i+1] == midi[i]: 
            i += 1
            l += 1
            if i+1 == len(midi): break
        i += 1
        d.append([p, l])
    return np.matrix(d)
```

```python
def realtime_pitch(stream, buffer_size):
    audiobuffer = stream.read(buffer_size)
    signal = np.frombuffer(audiobuffer, dtype=np.float32)
    bins_per_octave = 12
    cqt = librosa.cqt(signal, sr=sr, n_bins=60, bins_per_octave=bins_per_octave)
    log_cqt = librosa.amplitude_to_db(np.abs(cqt))
    return log_cqt.argmax(0)[1]+24
```

```python
# initialise pyaudio
p = pyaudio.PyAudio()

# open stream
buffer_size = 1024
pyaudio_format = pyaudio.paFloat32
n_channels = 1
sr = 44100

stream = p.open(format=pyaudio_format,
                channels=n_channels,
                rate=sr,
                input=True,
                frames_per_buffer=buffer_size)

print(sheet)

print("*** start recording")
num_frames = 0
index = 0
ratio = None

status = True
while True:
    try:
        if status == False: break
        ratio = None
        s = realtime_pitch(stream, buffer_size)
        while s!= sheet[0, 0]: 
            s = realtime_pitch(stream, buffer_size)
#             print(s)
        
        print("first note match: %d" %s)
            
        while index < sheet.shape[0]:
            print("index = %d"%index)
            l = 0
            while s == sheet[index, 0]: 
                l+=1
                s = realtime_pitch(stream, buffer_size)
            print("l = %d, pitch = %d"%(l, sheet[index, 0]))
            
            if ratio != None: 
                t = math.ceil(sheet[index, 1]*ratio/5)
#                 print("here", sheet[index, 1], ratio, t)
            else: t = math.ceil(l/5)
            
#             print("t = %d" %t)
            c = 0
            while c < t:                
                s = realtime_pitch(stream, buffer_size)
                c += 1                
#                 print("buff %d, pitch = %s"%(c, s))
                
                if index + 1 < sheet.shape[0] and s == sheet[index+1, 0]: 
#                     print("go to next note!")
                    break
                    
                if s == sheet[index, 0]:
                    while s == sheet[index, 0]: 
                        l+=1
                        s = realtime_pitch(stream, buffer_size)
                    c = 0
#                     print("new length = %d"%l)
            
            if ratio == None: 
                ratio = l/sheet[index, 1]
                index += 1
                print("initialize ratio to be", ratio)
            elif np.abs((ratio-l/sheet[index, 1])/ratio) < 0.25:
                if index == sheet.shape[0] -1: 
                    print("True")
                    plt.imshow(im)
                    status = False
                    break
                else: 
                    index += 1
            else: 
                if index == sheet.shape[0] -1 and l > 15: 
                    print("True")
                    plt.imshow(im)
                    status = False
                    break
                print("no way")
                print("l = ", l, "pitch", sheet[index, 0], l/sheet[index, 1])
                
                index = 0
                break
    except KeyboardInterrupt:
        print("*** Ctrl+C pressed, exiting")
        break


print("*** done recording")
stream.stop_stream()
stream.close()
p.terminate()
```
